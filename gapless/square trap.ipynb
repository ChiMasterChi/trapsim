Check to make sure this trap is confining.

```python
zp = 80e-6
yp = 0.
xwalk = np.arange(-200e-6, 200e-6, 1e-6)
pot = []
for xp in xwalk:
    pot.append( w.compute_pseudopot(xp, yp, zp) )

[fx, fy, fz] = w.compute_pseudopot_frequencies(0., 0., zp)
print fx*1e-6
print fy*1e-6
print fz*1e-6 # had better be 0
plot(array(xwalk)*1e6, array(pot))
xlabel('x distance (um)')
ylabel('Pseudopotential (meV)')
```

What happens if we walk diagonally across the center electrode?

```python
diagonal_walk = []
min = -200e-6
max = 200e-6
pot = []
zp = 20e-6
for j in arange(min, max, 1e-6):
    p = (j, j, 500e-6)
    diagonal_walk.append( j )
    pot.append( w.compute_pseudopot(j, j, zp) )
plot(array(diagonal_walk)*1e6, array(pot))
```
