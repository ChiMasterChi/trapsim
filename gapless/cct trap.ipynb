```python
from gapless import *
w = World(2*pi*35*1e6, d=1e-7)
Vrf = 100.
w.add_electrode('rf1', (-190e-6, -50e-6), (-50000e-6, 50000e-6), 'rf', Vrf)
w.add_electrode('rf2', (50e-6, 320e-6), (-50000e-6, 50000e-6), 'rf', Vrf)
zp = 107e-6
yp = 0.
xwalk = np.arange(-200e-6, 200e-6, 1e-6)
pot = []
xp = -12e-6
zp = 101e-6
for xp in xwalk:
    pot.append( w.compute_pseudopot(xp, yp, zp) )
print min(pot)
[fx, fy, fz] = w.compute_pseudopot_frequencies(-12e-6, 0., 107e-6)
print fx*1e-6
print fy*1e-6
print fz*1e-6
plot(array(xwalk)*1e-6, array(pot))
xlabel('x distance')
ylabel('Pseudopotential')
```

So far so good. The  planar trap gives deep confinement in the "x" direction (parallel to the trap plane). Moreover, the (degenerate) radial frequencies are about 3.4 MHz for a 35 MHz drive at 50 V amplitude. This seems to agree with what we see in the experiment, giving me confidence in the calculation. Now, what about the direction perpendicular to the trap? This is the "y" direction in the lab, but the "z" direction in this code.
